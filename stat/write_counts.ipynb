```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import math
```

```python
plt.rcParams["figure.dpi"] = 600

def run():
    mpl.rcParams['hatch.linewidth'] = 0.3
    count_uni = [5.85, .80, .80]
    xname = [r'IPS$^4$o', 'WiscSort', 'TSort']
    x = list(range(len(xname)))
    width = 0.5
    
    plt.figure(figsize=(2.4, 1.6))
    plt.bar(x, count_uni, width=width, edgecolor='#4584C3', color='white', linewidth=1.5)

    plt.xticks(x, xname, size=7)
    plt.yticks(range(0,7,1), size=7)
    plt.xlim(-0.5, 2.5)
    plt.ylim(0, 6)
    plt.ylabel(r'Write Count ($x10^8$)', size=7)
    plt.grid(linestyle='--', axis='y', alpha=0.3)
    plt.tight_layout()
    # plt.savefig('fig/write.pdf', format='pdf')

    plt.show()
```

```python
run()
```
